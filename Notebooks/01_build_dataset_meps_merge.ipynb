```python
## show the number of entire data instances
print("The number of entire data instances are:", condition.shape[0]) 

## Create a mask in which instances related to asthma are "True"
mask_Depress = condition['ICD10CDX'] == "F32"

print(mask_Depress)
```

```python
condition_DepressiveEpisode = condition[mask_Depress]

## show the number of selected data instances
print("The number of selected data instances are:", condition_DepressiveEpisode.shape[0])
```

```python
# Select duplicate rows except first occurrence based on all columns

## create a mask for duplicated instances
mask_duplicates = condition_DepressiveEpisode.duplicated()
duplicate_Rows = condition_DepressiveEpisode[mask_duplicates]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicate_Rows)
```

```python
# Select all duplicate rows based on one column

## create a mask for duplicated instances
mask_duplicates = condition_DepressiveEpisode.duplicated(subset=['DUPERSID'], keep=False)
duplicate_Rows = condition_DepressiveEpisode[mask_duplicates]

print("Duplicate Rows based on a single column are:", duplicate_Rows, sep='\n')
```

```python
# Look through this again

## Calculate the number of unique patients with Depressive episode

## create a mask for duplicated instances
## All duplicates except their first occurrence will be marked as True
mask_duplicates = condition_DepressiveEpisode.duplicated(subset=['DUPERSID'], keep="first")
print(mask_duplicates.value_counts())
print("The numnber of unique instances are:", condition_DepressiveEpisode.shape[0] - duplicate_Rows.shape[0])
```

```python
filtered_DepressiveEpisode = condition_DepressiveEpisode[~ mask_duplicates]
print(filtered_DepressiveEpisode.shape)
```

```python
filtered_consolidation = consolidation[["DUPERSID", "SEX"]]
filtered_consolidation
```

```python
joined = pd.merge(left = filtered_DepressiveEpisode, right = consolidation , how="inner", on='DUPERSID')

joined
```

```python
items_deletion = []
items_using = []

for item in joined.columns:
    if "_x" in item:
        items_using.append(item)
    elif "_y" in item:
        items_deletion.append(item)

print(items_deletion)
print(items_using)
```

```python
# drop the duplicate abbreviation column
joined = joined.drop(labels= items_deletion, axis= 1)

# change column name

for item in items_using:
    joined.rename(columns={item: item.replace("_x", "")}, inplace=True)
```
