```python


# assume agg_file_selected is already your DataFrame of interest

# 1. Mask all negative‐coded values to NaN
agg_file_selected = agg_file_selected.mask(agg_file_selected < 0, np.nan)

# 2. Count missing per column
missing_counts = agg_file_selected.isna().sum()

# 3. Keep only the columns that actually have missing values
missing_df = missing_counts[missing_counts > 0].reset_index()
missing_df.columns = ['variable', 'num_missing']

# 4. How many variables have any missing?
num_vars_with_missing = missing_df.shape[0]

# 5. Display results
print("Variables with missing values:\n", missing_df.head(), "\n")
print(f"Number of variables with missing values: {num_vars_with_missing}")
```

```python


# ----------------------------------------------------------------------------
# 0. Start from your pre-filtered DataFrame:
#    agg_file_selected = <your DataFrame of just the green-shaded vars>
df = agg_file_selected.copy()

# ----------------------------------------------------------------------------
# 1. Mask all MEPS “skip/refuse/not-applicable” codes (<0) to NaN
df = df.mask(df < 0, np.nan)

# ----------------------------------------------------------------------------
# 2. Impute the two major continuous fields by their column means
for col in ('AGEDIAG','ACCDNWRK'):
    if col in df:
        df[col].fillna(df[col].mean(), inplace=True)

# ----------------------------------------------------------------------------
# 3. Drop any redundant ID/index columns (if present)
for drop_col in ('CONDIDX',):
    if drop_col in df:
        df.drop(columns=[drop_col], inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)

# ----------------------------------------------------------------------------
# 4. Identify numeric columns for scaling
numeric_cols = df.select_dtypes(include='number').columns.tolist()

# 5. Z-score standardization
zscore_df = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()

# 6. Min–Max scaling
minmax_df = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())

# ----------------------------------------------------------------------------
# 7. Re-attach any ID columns you need (e.g. DUPERSID)
id_cols = [c for c in agg_file_selected.columns if not c in numeric_cols]
result  = pd.concat([agg_file_selected[id_cols], 
                     zscore_df.add_suffix('_std'), 
                     minmax_df.add_suffix('_mm')], axis=1)

# 8. Reorder: IDs first, then all _std, then all _mm
std_cols = [c for c in result if c.endswith('_std')]
mm_cols  = [c for c in result if c.endswith('_mm')]
result   = result[id_cols + std_cols + mm_cols]

# ----------------------------------------------------------------------------
# Now `result` holds your fully preprocessed features in memory.
# You can continue with:
#    • modeling (e.g. train_test_split, fit models)
#    • more feature engineering
#    • visualizations
#    • etc.
#
# Simply refer to `result` in whatever comes next.

print(result.head())
```

```python
### Noisy data 
```

```python
# -----------------------------------------------------------------------------
# 0. Start from your pre-filtered DataFrame:
#    agg_file_selected = <your DataFrame of just the green-shaded vars>
data_cleaned = agg_file_selected.copy()

# -----------------------------------------------------------------------------
# 1. Mask all MEPS “skip/refuse” codes (<0) to NaN
data_cleaned = data_cleaned.mask(data_cleaned < 0, np.nan)

# -----------------------------------------------------------------------------
# 2. Impute the two key continuous fields by their column means
for col in ('AGEDIAG','ACCDNWRK'):
    if col in data_cleaned:
        data_cleaned[col].fillna(data_cleaned[col].mean(), inplace=True)

# -----------------------------------------------------------------------------
# 3. Drop any redundant ID columns (if present)
if 'CONDIDX' in data_cleaned.columns:
    data_cleaned.drop(columns=['CONDIDX'], inplace=True)

# Reset the index
data_cleaned.reset_index(drop=True, inplace=True)

# -----------------------------------------------------------------------------
# 4. Define which columns are truly continuous
continuous = [
    'HHNUM','IPNUM','OPNUM','OBNUM','ERNUM','RXNUM',
    'FCSZ1231','RUSIZE31','RUSIZE42','RUSIZE53',
    'AGE31X','AGE42X','AGE53X','K6SUM42','PHQ242'
]
cont = [c for c in continuous if c in data_cleaned.columns]

# 5. Compute IQR boundaries
Q1  = data_cleaned[cont].quantile(0.25)
Q3  = data_cleaned[cont].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 6. Filter out rows with any outlier
mask = ~(
    (data_cleaned[cont] < lower_bound) |
    (data_cleaned[cont] > upper_bound)
).any(axis=1)

data_filtered = data_cleaned[mask].reset_index(drop=True)
print(f"Kept {len(data_filtered)} of {len(data_cleaned)} rows after outlier filtering.")

# -----------------------------------------------------------------------------
# 7. (Alternative) Winsorize instead of drop
data_winsorized = data_cleaned.copy()
data_winsorized[cont] = data_winsorized[cont].clip(lower=lower_bound, upper=upper_bound, axis=1)
print("Winsorization applied: outliers capped to 1.5×IQR bounds.")

# -----------------------------------------------------------------------------
# Next steps (as in Week 7):
#  - Use `data_filtered` (or `data_winsorized`) for your normalization and scaling steps
#  - Re-attach any ID columns (e.g. DUPERSID) if needed
#  - Proceed to modeling, clustering, or further feature engineering
```

```python
### Now that we’ve filtered out (or winsorized) the outliers, the very next step in the Week 7 pipeline is to normalize and scale those cleaned continuous features so they’re on comparable footing for any downstream analysis or modeling.

###Here’s how you’d pick up from either data_filtered (outlier‐removed) or data_winsorized and do the z-score + min–max steps, then reattach any ID columns:
```

```python
# 0. Choose your cleaned set:
#    use data_filtered        # if you dropped outliers
# or use data_winsorized      # if you capped them instead
clean = data_filtered.copy()

# 1. Identify numeric features to scale
#    (assumes all the green-shaded vars in clean are numeric or encoded)
numeric_cols = clean.select_dtypes(include='number').columns.tolist()

# 2. Z-score normalization
zscore = (clean[numeric_cols] - clean[numeric_cols].mean()) \
           / clean[numeric_cols].std()

# 3. Min–max scaling
minmax = (clean[numeric_cols] - clean[numeric_cols].min()) \
           / (clean[numeric_cols].max() - clean[numeric_cols].min())

# 4. Reattach identifier(s), e.g. if you have DUPERSID or CONDN still around:
id_cols = [c for c in clean.columns if c not in numeric_cols]
final_df = pd.concat([clean[id_cols], 
                      zscore.add_suffix('_std'), 
                      minmax.add_suffix('_mm')],
                     axis=1)

# 5. Reorder: IDs first, then all _std, then all _mm
std_cols = [c for c in final_df if c.endswith('_std')]
mm_cols  = [c for c in final_df if c.endswith('_mm')]
final_df = final_df[id_cols + std_cols + mm_cols]

# Now final_df holds your fully preprocessed, scaled features.
```
